# Scatter plots

```python
import numpy as np
import csv
from matplotlib import pyplot as plt

filepath = '/workspaces/codespaces-jupyter/data/data8.csv'
```

```python
# csv_reader = {'view_count': '8036001', 'likes': '324742', 'ratio': '96.91'}
```

```python
views = np.array([], dtype='int64')
likes = np.array([], dtype='int64')
ratio = np.array([], dtype='float')
with open(filepath) as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        views = np.concatenate((views, np.array([row['view_count']], dtype='int64')))
        likes = np.concatenate((likes, np.array([row['likes']], dtype='int64')))
        ratio = np.concatenate((ratio, np.array([row['ratio']], dtype='float')))
```

- s (size)
- c (color)
- cmap (color theme)
- marker
- colorbar() -> plt.colorbar().set_label('Like-Dislike ratio')
- plt.xscale()
- plt.yscale()

```python
color = np.random.randint(25,100, size=200)
plt.scatter(views, likes, s=50, alpha=0.5, c=ratio, cmap='summer', edgecolor='none', linewidth=1, marker='o')

plt.colorbar().set_label('Like-Dislike ratio')

plt.xscale('log')
plt.yscale('log')

plt.title("Trending YouTube Videos - 31/05/2019")
plt.xlabel("View Count")
plt.ylabel("Total Likes")
plt.tight_layout()
plt.show()
plt.close()
```
