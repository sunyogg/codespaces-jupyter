# CSV Handling

Importing the csv module

```python
import csv
```

```python
filepath = "/workspaces/codespaces-jupyter/data/names.csv"
```

This is something that we can do without using the __csv__ module. But it is kinda messy

```python
fname = []
lname = []
email = []
with open(filepath, 'r') as csv_file:
    flag = True
    next(csv_file)
    while(flag):
        try:
            result = next(csv_file).split(',')
            fname.append(result[0])
            lname.append(result[1])
            email.append(result[2].split('\n')[0])
        except StopIteration:
            flag = False
```

### Reading the CSV file

__csv.reader()__ creates an object by taking the csv file as input and it makes reading from a csv file a whole lot easier. 

```python
with open(filepath, 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    next(csv_reader)
    for line in csv_reader:
        print(line[-1])
        
```

### Writing on a file using CSV module

```python
new_filepath = "/workspaces/codespaces-jupyter/data/new_names.csv"
```

```python
with open(filepath, 'r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')

    with open(new_filepath, 'w') as new_file:
        csv_writer = csv.writer(new_file, delimiter='\t')
        
        for line in csv_reader:
            csv_writer.writerow(line[0:2])
```

### Dictionary Reader

- csv.DictReader() creates an object which stores the csv data as dictionaries where __column names__ are the __keys__ and __rows__ are __values__ of dictionary. 
- It makes accessing values of csv file even more easier and similar to how we access __values__ using the __keys__ in a dictionary.
- Makes the code lot more readable as we are not writing the indexes of the lists like before instead we using keys to get the values associated to those keys.  

```python
with open(filepath, 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    print(type(csv_reader))
    for line in csv_reader:
        print(line)
```

### Dictionary Writer

- this method __csv_writer.writerow(row)__ only writes the __values__ of the dictionary which is what we want. 
- The inner working of all this must be pretty complex. 
- It is probably comparing the fieldnames of csv_reader and csv_writer and then simply putting the values as the record.
- That's why it throws error if fielname contain different column names compared to columns in csv_reader.

```python
with open(filepath, 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file, delimiter=',')
    fieldnames = ['first_name', 'last_name', 'email']
    with open(new_filepath, 'w') as new_csv_file:
        csv_writer = csv.DictWriter(new_csv_file, fieldnames=fieldnames, delimiter='\t')
        csv_writer.writeheader()
        for row in csv_reader:
            csv_writer.writerow(row)
```
