# while loop

We use while loop when we want to loop through a task but we don't know for how many times we are going to repeat that task.

```python
i = 0
while (i < 5):
    print(f"'i' is {i}")
    i += 1
```

## while else loop

```python
i = 0
flag = False
while (i < 10):
    print(i)
    i += 1

    if (i == 9 and flag):
        break
else:
    print("This else part will only be executed if the loop has ran all it's cycle without being interrupted by the 'break' statement")
```

## break and continue statement

```python
i = 0
while (i < 20):
    i += 1
    if (i % 2 == 0):
        print(f"{i} is even")
    else:
        # i is odd
        continue
    if (i == 18):
        break
```
