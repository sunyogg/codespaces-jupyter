## if elif else

```python
age = 51
if (age < 0):
    print("Age can't be negative")
elif (age > 0 and age <= 2):
    print("You're a baby!")
elif (age > 2 and age <= 12):
    print("You're a child!")
elif (age > 12 and age <= 19):
    print("Happy teen-ing!")
elif (age > 19 and age <= 35):
    print("You're young adult")
elif (age > 35 and age <=50):
    print("You're a middle-aged adult")
else:
    print("You're getting old")
```

## Nested if else

```python
num = int(input("Enter a number between 0 and 40: "))
if num < 40:
    if num >= 30:
        print("It lies between 30 and 40")
    else:
        if (num >= 20):
            print("It lies between 20 and 30")
        else:
            if (num >= 10):
                print("It lies between 10 and 20")
            else:
                if (num >= 0):
                    print("It lies between 0 and 10")
                else:
                    print("It's a negative number")
else:
    print("Number is greater than 40")
        
```

## match statement
- Better to use match statement instead of many if elif else statement.
- Don't need __break__ statement unlike other programming language.

```python
day = int(input("Enter day number: "))
match (day):
    case (1):
        print("monday")
    case (2):
        print("tuesday")
    case (3):
        print("wednesday")
    case (4):
        print("thursday")
    case (5):
        print("friday")
    case (6):
        print("saturday")
    case (7):
        print("sunday")
```
