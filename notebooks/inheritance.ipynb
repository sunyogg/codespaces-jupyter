# Inheritance

- We don't always have to create a class from beginning, if the class we're writing is a specialised version of the class we have created then we can use inheritance and reuse the code. 
- When one class inherits from another, it takes on the attributes and methods of the first class.
- The original class is called the __parent class__, and the new class is the __child class__.
- The child class can inherit any or all of the attributes and methods of its parent class, but it’s also free to define new attributes and methods of its own.
- __Is a__ relationship between classes.
    - __ElectricCar__ is a __Car__.
    - __Woman__ is a __Human__.
    - __Admin__ is a __User__.

#### Without __init__\() in child class
- With no __init__\() function in Child class we can simply say that __Child__ class === __Parent__ class, with almost no difference.
- When you add the __init__\() function, the child class will no longer inherit the parent's attributes.

```python
class Parent:
    def __init__(self, name, age, pay):
        self.name = name.title()
        self.age = age 
        self.pay = pay

    def describe(self):
        print(f"Name -> {self.name}")
        print(f"Age -> {self.age}")
        print(f"Pay -> {self.pay}/-")

    def increment_pay_by_percent(self, percent):
        self.pay += self.pay * (percent / 100)

# Child class INHERIT from Parent class.
class Child(Parent):
    # It'd inherit all the atttributes and methods. 
    pass

child = Child("sunyog", 22, 100)
child.describe()
```

#### With __init__\() in child class
- We'd need to initialize attributes that are taken as input in the child's __init__\() method.
- Now since a child class can have attribute of parents class as well as attributes unique to it.
    - Instead of initializing the attributes again that are __common__ in both Parent and Child class we can use __super()__ to initialize the common attribute then we'd only need to initialize the __unique attributes__.
    - Below you can see we are using super() to initialize common attributes like __name, age__ and __pay__ and whereas we are explicitly initializing the unique attributes like __education__.

```python
class Parent:
    def __init__(self, name, age, pay):
        self.name = name.title()
        self.age = age 
        self.pay = pay

    def describe(self):
        print(f"Name -> {self.name}")
        print(f"Age -> {self.age}")
        print(f"Pay -> {self.pay}/-")

class Child(Parent):
    def __init__(self, name, age, pay, education):
        # Now we also have to initialize the attributes just like the parent class.
        # And we can also initialize the unique attributes for the child class.
        self.name = name
        self.age = age
        self.pay = pay
        self.education = education

kid = Child("sunyog", 22, 2000, "college")
```

#### super().__init__\()

- Instead of initializing the attributes again that are common in both Parent and Child class we can use super() to initialize the common attribute then we'd only need to initialize the unique attributes.
- Below you can see, we are using super() to initialize common attributes like __name, age__ and __pay__ and whereas we are explicitly initializing the unique attributes like education.
- The super() function is used to give access to methods and properties of a parent or sibling class. It returns an object that represents the parent class.
- But if you're using super().__init__\(), then you have to make sure that you're taking all the attributes of parent class as child class input, this is because of us calling the parent class constructor and the parent constructor will require the number of argument that were initialized during it's creating in the parent class.
- So if you are using super().__init__\() then you're losing the ability to
    - Initialize __some of the variable of parent__ class and some other unique variables.

```python
class Parent:
    def __init__(self, name, age, pay):
        self.name = name.title()
        self.age = age 
        self.pay = pay

    def describe(self):
        print(f"Name -> {self.name}")
        print(f"Age -> {self.age}")
        print(f"Pay -> {self.pay}/-")

class Child(Parent):
    def __init__(self, name, age, pay, education):
        # We can use super() to call the constructor of parent class to initialze the common attributes in order to save some time. 
        super().__init__(name, age, pay)
        self.education = education
        
# print(child.name) AttributeError: 'Child' object has no attribute 'name'

kid = Child("sunyog", 22, 2000, "college")
```

### Method Overriding
- A method in Parent class can also exist in Child class while also performing a different function.

```python
class Rectangle:
    def __init__(self, length, breadth):
        """This method will initialize a rectangle"""
        self.length = length
        self.breadth = breadth

    def get_length(self):
        """This method will return the length of a rectangle"""
        return self.length

    def get_breadth(self):
        """This method will return the breadth of a rectangle"""
        return self.breadth
        
    def get_area(self):
        """This method will return the area of a rectangle"""
        area = self.length * self.breadth
        return area

class Circle(Rectangle):
    def __init__(self, radius):
        """This method initializes a circle"""
        self.radius = radius

    def get_radius(self):
        """This method will return the radius of the circle"""
        return self.radius

    # Method overriden
    def get_length(self):
        """This method will return the perimeter of the circle"""
        perimeter = 2 * 3.14159 * self.radius
        return perimeter

    # Method overriden
    def get_area(self):
        """This method will return the area of the circle"""
        area = 3.14159 * self.radius * self.radius
        return area

circle = Circle(5)
circle.get_area()
circle.get_length()
```

### Multiple Inheritance
- When a child class has multiple Parents.
- When a class inherits it's properties and methods from more than one Class.
- Not supported in many languages cause it __causes diamond problem__.
- The Problem - When C is inheriting from A and B. And A and B contains method of same name, then if we call that method using the the object of C. The interpreter will get confuse if that method is not present in C, it'll have to decided whether it should call that method from B or A.
- Study about [MRO (Method Resolution Order)](https://youtu.be/PSMYqfMP3Cs?si=e64QfPb0vWGAGANe) and C3 linearization algorithm __in future__.
- Python solves this problem by putting preference in the sequence of in heritance.
    - if __class C(B, A)__ -> then it'll __first search in B and then in A__.
    - if __class C(A, B)__ -> then it'll __first search in A and then in B__.

```python
class A:
    def __init__(self):
        pass
        
    def greet(self):
        print("Hello A, How are you doing?")


class B(A):
    def __init__(self):
        pass
        
    def greet(self):
        print("Hello B, How are you doing?")


class C(A):
    def __init__(self):
        pass
        
    def greet(self):
        print("Hello C, How are you doing?")


class E(C, B):
    pass
e = E()
e.greet()

```
