### Modifying tuples

```python
#Tuples are immutable and cannot be modified
tup[0] = "cherries"

#But we can change the object that tuple variable is refrencing to
tup = [7, 8, 9, "tomato", "potato", "onions"]
print(tup)
```

### Tuple unpacking

```python

# Python provides us with an elegant syntax for unpacking the individual elements of a sequence as separate variables. 
# We call this unpacking.
# However, the number of variables on the left-hand side must match the number of elements in the tuple on the right-hand side.
name, hobbies, parents = ("sunyog", ['gaming', 'walking', 'coding', 'singing'], {"mother":"Kiran", "father":"Mahesh"})
print(name)
print(hobbies)
print(parents)

# If the number of variables is less than the number of values, 
# you can add an * to the variable name and the values will be assigned to the variable as a list:
name, hobbies, *games = ("sunyog", ['gaming', 'walking', 'coding', 'singing'], "gta", "mafia", "cod", "rdr")
print(name)
print(hobbies)
print(games)
```
