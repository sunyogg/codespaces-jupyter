# Pie chart

- Works fine when you have less than 5 items.

__Attributes:__
- labels
- colors
- wedgeprops
- shadow
- counterclock
- frame
- rotatelabels
- explode
- startangle
- radius
- __autopct__

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
values = [35, 73, 17]
labels = ["Mac", "Windows", "Linux"]
plt.pie(values, labels = labels, colors=["#fa5252", "#D1D7DC", "#dee2e6"], shadow=True, startangle=260, radius=1.2, counterclock=True, 
        frame=False, rotatelabels=False, explode=[0.15, 0, 0],
        wedgeprops={"edgecolor":"black", "linewidth":1, "ls":"-", "alpha":1 })
plt.show()
plt.close()
```

```python
langs = ['HTML/CSS', 'Java', 'JavaScript', 'Python', 'C++',]
values = [55466, 35917, 59219, 36443, 20523]
```

```python
plt.pie(values, labels = langs, explode=[0, 0, 0, 0.15, 0], shadow=True, startangle=105, 
        colors=["#fa5252", "#12b886", "#9775fa", "#228be6", "#f59f00"], wedgeprops={"edgecolor":"black"}, autopct="%1.1f%%")
plt.show()
plt.close()
```
