## Merge

```python
df_left = pd.DataFrame({
    "id" : np.arange(7, 12, dtype='int8'),
    "name" : ["Rahul", "Ravi", "Mahesh", "Sonu", "Prasad"],
    "age" : [12, 27, 54, 28, 79]
})

df_right = pd.DataFrame({
    "id" : np.arange(7, 13, dtype='int8'),
    "occupation" : ['Student', 'Electrician', 'Farmer', 'Engineer', 'Retired', 'Military'],
    "location" : ['Mumbai', "Delhi", 'Jabalpur', "Pune", "Bhopal", 'Kashmir'],
    "married" : [False, True, True, False, True, True]
})
```

```python
df_merged = pd.merge(df_left, df_right, on="id")
df_merged
```

## Grouping
- __Splitting__ the data into groups based on some criteria
- __Applying__ a function to each group independently
- __Combining__ the results into a data structure



```python
df.head()
```

With sort=False the order among group-keys follows the order of appearance of the keys in the original dataframe

```python
grouped = df.groupby("gender", observed=True, sort=False).mean('pay')
grouped.head()
```

```python
# grouped = df.groupby("gender", observed=True, sort=False)[['experience', 'pay']].mean('pay')
grouped = df.groupby("gender", observed=True, sort=False).mean('pay')[['experience', 'pay']]
grouped
```

__We only had unique values but our indexes were duplicates. So data was grouped together on the basis of same indexes.__<br></br>
If a non-unique index is used as the group key in a groupby operation, all values for the same index value will be considered to be in one group and thus the output of aggregation functions will only contain unique index values

```python
index = 1, 2, 3, 1, 2, 3, np.nan
data = pd.DataFrame([11, 12, 13, 14, 15, 16, 17], index = index, columns=['values'])
grouped = data.groupby('values', level=0, sort=True, dropna=True).sum()
grouped
```
