## 'return' in function

We can use 'return' statement to make the function 'return' a value. 
Even functions without a return statement do return a value, albeit a rather boring one. This value is called None (it’s a built-in name).

We can use the returned value to pass it onto another function or store it in a variable.

```python
def sum(x, y):
    """This function returns the sum of 2 numbers taken as arguments"""
    result  = x + y
    return result

# Notice how we are storing the return value of a function.
summation = sum(2, 3)
```

Function overloading: A method where you can create multiple functions of same name and you can differentiate them by:<br> 1) Changing the type of parameter. <br>2) Changing the sequence of parameter. <br>3) Changing the number of parameters. <br>And compiler will decide in run time which function to call.

## POSITIONAL and KEYWORD and DEFAULT arguments

### Positional arguments
- Arguments are passed in the order of parameters. The order defined in the order function declaration.
- Order of values cannot be changed to avoid the unexpected output.

```python
def fun(first_arg, second_arg, third_arg):
    print(f"first_arg is {first_arg}")
    print(f"second_arg is {second_arg}")
    print(f"third_arg is {third_arg}", "\n")

fun(1, 2, 3)
fun("sunyog", "lodhi", "brown")
```

__Note__ - We have to make sure that we are passing the arguments in the order that matches with the order of parameters defined during the definition of the function.

```python
def nameAge(name, age):
    print(f"Hi, I am {name}")
    print(f"My age is {age}", "\n")
nameAge("sunyog", 22)
nameAge(22, "sunyog")
```

### Keyword arguments

- Parameter Names are used to pass the argument during the function call.
- Order of parameter Names can be changed to pass the argument.
- Whether a function is keyworded or positional depends upon how you're calling the function.

```python
def nameAge(name, age):
    print(f"Hi, I am {name}")
    print(f"My age is {age}", "\n")

# This is a KEYWORD argument
nameAge(name = "sunyog",  age = 22)
nameAge(age = 22, name = "sunyog")
```

### Default arguments
- Default values indicate that the function argument will take that value if no argument value is passed during the function call.

```python
def nameAge(name = "sunyog", age = 22):
    print(f"Hi, I am {name}")
    print(f"My age is {age}", "\n")

#If we don't pass arguments then default values will be used which are initialized while defining the function. 
nameAge()
#positional argument
nameAge("suyog", 23)
#keyword argument
nameAge(age = 26, name = "prasoon")
```

## Variable Number of arguements in a function

### *args
- __variable number of non-keword arguments__
- The special syntax __*args__ in function definitions in Python is used to pass a variable number of arguments to a function.
- It is used to pass a __non-keyworded__, __variable-length__ argument list.
- Those variable number of arguments in __*args__ are stored inside __tuple__, you can check this by printing the __type(args)__. And hence these __*args__ will follow all the properties of __tuple__.

```python
def fav_food(*args):
    print(args)
    print(f"First element is {args[0]}")

fav_food("Pizza", "Burger", "Noodles")
```

Since those variable number of arguments in *args are stored inside tuple, hence these *args will follow all the __properties__ of __tuple__.

```python
def mod_fav_food(*args):
    #Remember we cannot modify a tuple
    args[0] = "mangos"
    print(args)

mod_fav_food("Pizza", "Burger", "Noodles")
```

### **kwargs

- __Variable number of keyword arguments__
- The special syntax __**kwargs__ in function definitions in Python is used to pass a variable number of arguments to a function.
- It is used to pass a __keyworded__, __variable-length__ argument list.
- Those variable number of arguments in __*args__ are stored inside __dictionary__, you can check this by printing the __type(kwargs)__. And hence these __**kwargs__ will follow all the properties of __dictionaries__.
- A __keyword argument__ is where you provide a name to the variable as you pass it into the function.

```python
def fav_food(**kwargs):
    print(type(kwargs), "\n")
    print(kwargs, "\n")
    for key, value in kwargs.items():
        print(f"{key} -> {value}")

fav_food(fname = "sunyog", lname = "lodhi", age = 22, color = "brown")
```

Since those variable number of arguments in __**kwargs__ are stored inside tuple, hence these *args will follow all the __properties__ of __dictionaries__.

```python
def mod_fav_food(**kwargs):
    kwargs["age"] = 32
    print(kwargs)

mod_fav_food(fname = "sunyog", lname = "lodhi", age = 22, color = "brown")
```

## Python is 'Call by Object Reference'
The two most widely known and easy to understand approaches to parameter passing amongst programming languages are __pass-by-reference__ and __pass-by-value__. Unfortunately, Python is __“pass-by-object-reference”__, of which it is often said:<br>__“Object references are passed by value.”__<br><br>
Whether a parameter will be passed by __reference__ or passed by __value__ depends on whether the parameter is __mutable__ or __immutable__.<br>
__if__ object is __immutable__:<br>
>   use pass by __value__
> 
__elif__ object is __mutable__:<br>
> use pass by __reference__
>

#### Immutable objects
- (Integers, Float, String, Tuple)
- __Pass By Value__
- Pass by value -> copy of value is being passed.
- A copy of value of var (copy of 10) is being passed on the function.
- __var__ in "__var__ = 10" is different from __var__ in "def scope(__var__)"
- A new object is created in the memory because integer objects are immutable.

```python


def scope(var):
    var = var + 5
    print(var, id(var))
    return var
var = 10
print(var, id(var))
scope(var)
print(var, id(var))
```

#### Mutable objects
- (list, dictionaries)
- Pass by reference (if you're modifying the list and not assigning a new list)
- pass by reference -> address or reference is being passed as the parameter.
- A new object is not created in the memory because list objects are mutable (modifiable). It simply adds new element to the same object.

```python
fruits = ["banana", "mango", "orange", "apple"]

def mod_list(array):
    array[0] = "strawberry"
    array.append("pomegranate")
    print(f"{id(array)} -> {array}")

# You can see how the id() of all the objects are same. Which indicates that they are stored at the same location.
print(f"{id(fruits)} -> {fruits}")
mod_list(fruits)
print(f"{id(fruits)} -> {fruits}")
```
