# Lists

Collection of similar of differnt types of data.
Enclosed in square brakcets separated by commas.
List allows you to store set of information in one place 

### Creating List

```python
squares = [1, 4, 9, 16, 25, 36]
print(squares)
```

### Addition operators in list

```python
squares = squares + [64, 81, 100, 121]
print(squares)
```

### Accessing and Modifying List elements

```python
#Accessing elements of list
print(squares[0])
print(squares[-1])

#Modifying elements of list.
squares[-1] = 49
print(squares)
```

### Deleting List elements

```python
del squares[0]
```

### List Slicing

```python
print(squares[:])
print(squares[0:3])
print(squares[0:7:2])
print(squares[0:-4])
```

### Checking if element present in list

```python
details = ["sunyog", "lodhi", 22, "brown", 66]
boolean = "lodhi" in details

if ("suyog" in details):
    print("suyog is in list")
else:
    print("suyog does not exist in list")
print(boolean)
```

### Looping through Lists

```python
for item in squares:
    result = item * 10
    print(item)
```

### Nested Lists

```python
nested = [[2, 4, 6],["apple", "mango", "banana"],["summer", "winter", "autumn"]]
accessing_nested_list  = nested[1][0]
print(accessing_nested_list)
```

### List comprehension

```python
numbers = [x for x in range(0,12) if x % 2 == 0]
print(numbers)
```

### List methods

```python
#count the length of list
length = len(numbers)

# Append an element at the last of an list.
numbers.append(6)

#Insert an element at any specified index. 
numbers.insert(0, -2)

# remove the last element of list.
# this method also returns the popped elements which then can be stored on any variable.
popped = numbers.pop();

# Removes the first item with the specified value
numbers.remove(6);

#Returns the copy of the list.
another_list = numbers.copy()

#Returns the index of the first element with the specified value
numbers.index(2)

#Reverses the order of the list
numbers.reverse()

#Sorts the list
numbers.sort(reverse = True)

print(numbers)

# Removes all the elements from the list
numbers.clear()

# Using range function and list to create a list
new_list = list(range(0, 10))
```
