# Plotting time series data in matplotlib

```python
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
from datetime import datetime, timedelta
import numpy as np
import csv

filepath = "/workspaces/codespaces-jupyter/data/data9.csv"
```

```python
price_date = np.array([], dtype='datetime64')
price_close = np.array([], dtype='float64')

with open(filepath) as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        price_date = np.concatenate((price_date, np.array([row['Date']], dtype='datetime64')))
        price_close = np.concatenate((price_close, np.array([row['Close']], dtype='float64')))


price_date = (np.sort(price_date))


plt.plot_date(price_date, price_close, linestyle = '-')
plt.gcf().autofmt_xdate()

plt.grid(True)
plt.title("Bitcoin Prices over a week")
plt.ylabel("Price")
plt.xlabel("Days")
plt.show()
plt.close()
```
