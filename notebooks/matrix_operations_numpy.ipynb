```python
import numpy as np
```

# Matrix Operations

#### Arithmetic operations

```python
mat_1 = np.array([
    [2, 5, 2],
    [3, 1, 0],
    [1, 4, 3]
])

mat_2 = np.array([
    [1, 4, 2],
    [2, 5, 6],
    [3, 2, 3]
])

print(mat_1, "\n")
print(mat_2, "\n")
```

```python
# matrix addition

result = np.add(mat_1, mat_2)


print(result)
```

```python
# matrix subtraction

result = np.subtract(mat_1, mat_2) # for b - a 

print(result)
```

```python
# matrix division

result = np.divide(mat_1, mat_2) # for a / b

print(result)
```

```python
# matrix scaler multiplication

result = 5 * mat_1

print(result)
```

```python
# element wise matrix multiplication (Aij * Bij)

result = np.multiply(mat_1, mat_2)

print(result)
```

```python
# matrix vector multiplication | MATRIX MULTIPLICATION
# rows of mat_1 should be equal to col of mat_2

print("Using np.dot()")
result = np.dot(mat_1, mat_2)

print(result, "\n")

print("Using np.matmul()")
print(np.matmul(mat_1, mat_2))
```

#### Determinant of a matrix

```python
print(mat_2, "\n")

result = np.linalg.det(mat_2)

print(result)
```

#### Transpose of matrix
- Rows -> Cols
- Cols -> Rows

```python
print(mat_1, "\n")

print("Using mat_1.T")
result = mat_1.T

print(result, "\n")

print("Using np.transpose(mat_1)")
print(np.transpose(mat_1))
```

#### Inverse of a matrix

```python
print(mat_1, "\n")

result = np.linalg.inv(mat_1)

print(result)
```
