# Object Oriented Programming in python

__What really is object oriented programming?__

- Type of methodology / way to write a program, which made real world __problems solvable by programs__ really __easier__.
- In which programs are written as classes and objects.
- Classes are like blueprint which can be used to create identical objects.
    - A __Car__ class can be used to create objects (cars) like __Tesla__, __Audi__, __Hundai__, __Tata__.
    - An __Employee__ class can be used to model employees as objects where each object will be a real life employee with some attributes like pay, email, experience, skills and methods like skill and work. 
- Objects can be think of any entity just like in real life which has some attributes and behavior.
    - __Height, Weight, Color, Age__ can be considered as an __attribute__ of 'human object'.
    - __Run, Walk, Sleep, Eat, Pee, Fight, Jump__ can be considered as __behavior__ of 'human object'.

__Why was Object oriented programming developed or needed?__

- Procedural programming was challenging to model real-world problems effectively.
- Reusability of code increased due to paradigms like inheritance and method overriding.
- Features like encapsulation allowed data and the methods that operate on the data into a single unit(object). This was beneficial in data protection and modularity and allowed programmer to control the access of internal details of an object. 
- OOP provides abstraction, allowing developers to focus on the essential features of an object while hiding unnecessary details.
- Made modelling real world problems easier. As it allowed programmers to create object which could have attributes and behavior just like any real world object. <br><br>
- Imagine modelling a __Library management system__ with and without classes. It surely is going to be really difficult to create something this complex without using object oriented programming:
    - We can have main __Library__ class which will act as a __Parent__ class to __Dean__, __Librarian__, __Student__ and __other__ classes.
    - We could assign different responsiblity and access control to different classes, we can use attributes of the object to authenticate different conditions
        - __Student__ of some __semester, branch, cgpa__ (attributes) can only __issue__ and __return__ (methods) a book
        - A __librarian__ having __experience, pay, skills__ (attributes) can __give__ and __take__ books back from student and s/he can also __buy__ (methods) new books for the library.
        - A dean having __pay, experience__ (attributes)might have full control of the library:
            - A __student to ban__.
            - Assign a __new librarian__.
            - __Authorize__ the types of books in the library.
            - __Monitor and control__ expenses used by the library. (methods).

#### Let's create a Human class (Become GOD!)
- So here __Human__ class will act as __blueprint__ to __create individual humans__.
- These individual humans will share common attributes like __name, color, height, weight, age, role.__
- These individual humans will also share common behavior like __sleep, eat, run, study.__

```python
class Human:

    population = 0

    def __init__(self, name, color, height, weight, age, role):
        """Initialize the attribute of human object."""
        print(self)
        self.name = name.title()
        self.color = color.title()
        self.height = height
        self.weight = weight
        self.age = age
        self.role = role.title()
        # a default attribute
        self.max_age = 120
        Human.population += 1

    def sleep(self):
        """This method replicates the sleeping behavior of human beings."""
        print(f"{self.name} is now sleeping.")
        
    def eat(self):
        """This method replicates the eating behavior of human beings."""
        print(f"{self.name} is hungry, give him something to eat.")
        
    def run(self):
        """This method replicates the running behavior of human beings."""
        print(f"Maybe it's an emergency, that's why {self.name} is running.")
        
    def study(self):
        """This method replicates the studying behavior of human beings."""
        print(f"{self.name}! It's time to study and do something good in life.")

    def bio_data(self):
        """This method prints bio data of a human."""
        print(f"Name -> {self.name}")
        print(f"Color -> {self.color}")
        print(f"Height -> {self.height}")
        print(f"Weight -> {self.weight}")
        print(f"Age -> {self.age}")
        print(f"Role -> {self.role}")
        
    def check_vote_eligibility(self):
        """This method will return true or false depending on whether a human is eligible for casting vote or not."""
        if (self.age < 18):
            print(f"{self.name}, you're not eligible to caste vote")
            return False
        else :
            print(f"{self.name}, you're eligibile to cast vote")
            return True

    def life_phase(self):
        if self.age < 2:
            print("You are Baby.")
        elif 2 <= self.age < 4:
            print("You are Toddler.")
        elif 4 <= self.age < 13:
            print("You are a kid.")
        elif 13 <= self.age < 20:
            print("You are a Teenager.")
        elif 20 <= self.age < 65:
            print("You are an Adult.")
        else:
            print("You are Elder.")


#Instantiation - Creating instance or object.
amrit = Human("Amrit Pal", "brown", 6.1, 73.8, 28, "Software developer")
ravi = Human("Ravi Patel", "brown", 5.8, 63.2, 24, "Student")
jeff = Human("Jeff luke", "white", 5.11, 68.2, 26, "Teacher")
carl = Human("Carl Johnson", "black", 5.7, 58.7, 17, "Student")
krishna = Human("Krishna Sukhdev", "brown", 1.1, 10.6, 1, "Baby")

ravi.bio_data()
#Human.bio_data(ravi)
krishna.life_phase()
```

## object (instance)
- Object is any entity that may be used to represent real world objects like living and non-living things.
- An object has:
    - __attributes__ - Features and characteristics of the object.
    - __methods__ - behavior of the object describing how it'll behave.

## attributes
- attributes == variables == data
- __instance variables/attributes__: used for data that is unique to each instance
- attributes are characteristics or features of any object.
- For example, attributes of a __human__ can be __age, weight, height, color, name, etc__.
- attributes of a __car__ can be it's __average, color, type, tank size, odometer reading, etc.__
- attributes can be defined without being passed in as parameters. These attributes can be defined in the __init__() method, where they are assigned a default value.

```python
class Human:
    def __init__(self, name, color, height, age):
        """Initialize the attribute of human object."""
        print(self)
        self.name = name.title()
        self.color = color.title()
        self.height = height
        self.age = age
        # a default attribute
        self.max_age = 120
    #--snip--
```

### class variable
- also known as __static variable__ and __class attributes.__
- When there are __variables or attributes__ that must be __common to each and every instance__ of a class.
- We call such variables as class variables.
- and they are not defined inside a constructor or any method but outside of the methods.
- For Example:
    - population is an attribute that is common for every human in this earth and is only incremented or decremented when a human is born or die.
    - Number of tyres of car is a class variable because every car has 4 tyres.
- They can be accessed using __self.__ or __Human.__ prefix (here).
- __self.population__ gives power to modify the class variable for a specific instance, so in this case __population might be different for different instances__, which doesn't make sense here, but imagine an Employee class having __annual_raise = 1.05__ as a class variable, now in this case depending on the employee's position and progress raise might be difference for different employee.
- __I also think that making a class variable unique for a instance might just make it a instance variable for that instance.__

```python
class Human:
    # A class variable
    population = 0

    def __init__(self, name, color, height, age):
        self.name = name
        self.color = color
        self.height = height 
        self.age = age
        Human.population += 1
    #--snip--
```
