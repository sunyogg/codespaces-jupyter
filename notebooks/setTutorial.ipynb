# Set

```python
# UNORDERED, 
# UNCHANGEABLE
# UNINDEXED
# No duplicate members.
```

### Creating set

```python
# Set only contains unique elements.
# And it'll automatically remove any duplicates. 
# elements are separated by commas and enclosed in curly braces. 
unique_nums = {1, 2, 3, 4, 5, 6, 6, 2, 3}
print(unique_nums);
```

### Set comprehension

```python
newSet = {x**2 for x in range(10) if x%2 == 0}
print(newSet)
```

### TRUE/FALSE and 1/0 in set

```python
#True and 1 are considered duplicates, similarly 0 and False are considered duplicates.
thisset = {True, 1, 0, False}
print(thisset)
```

### Accessing set

```python
# You cannot access items in a set by referring to an index or a key.
# But you can loop through the set items using a for loop.
for x in unique_nums:
    print(x)
print(6 in unique_nums)
```

### Modifying set

```python
# Once a set is created, you CANNOT CHANGE its items, but you can ADD or REMOVE  new items.

#Either we can use add() to add an element in the set
unique_nums.add(8)
unique_nums.add(11)
unique_nums.add(True)


# We can also insert any iterable by using the update() method.
thisset = {"apple", "banana", "mango"}
unique_nums.update(thisset)
print(unique_nums)
```

### Deleting Set element

```python
#remove() will remove the element and raise the error if the element does not exist. 
unique_nums.remove("apple");

#discard() will remove the element and won't raise the error in any situation. 
unique_nums.discard("apple");

#pop() will remove any element from the set and hence we shouldn't use this method. 
unique_nums.pop()

#clear() will empty the whole set
unique_nums.clear()
print(unique_nums)
```

### Looping through set

```python
for item in unique_nums:
    print(item)
```

### Set methods

```python

# add any element in the set
unique_nums.add("apple")

# add any iterable in the set
unique_nums.update([1, 2, 3])

# remove specified element and throw error if not exist
unique_nums.remove("apple")

# remove specified element don't throw error at all
unique_nums.discard("apple")

# remove any random element from the set
#unique_nums.pop()

# empty the whole set
unique_nums.clear()

# returns the copy of the set
unique_nums.copy()
```

### Set operation specific methods

```python
unique_nums = {1, 2, 3, 4, 4}
thisset = {3, 4, 5, 6, 4, 5}
print(f"unique_set -> {unique_nums}")
print(f"thisset -> {thisset}")

#Return a set containing the union of sets
union = unique_nums.union(thisset)

#Returns a set, that is the intersection of two or more sets
intersection = unique_nums.intersection(thisset)

#Returns a set containing the difference between two or more sets
differnce = unique_nums.difference(thisset)

#Return a set that contains all items from both sets, except items that are present in both sets:
symmetric_differnce = unique_nums.symmetric_difference(thisset)


#incase we also want to update the set on which we are applying set operation

#Removes the items in this set that are not present in other, specified set(s)
unique_nums.intersection_update(thisset)

#Removes the items in this set that are also included in another, specified set
unique_nums.difference_update(thisset)

#nserts the symmetric differences from this set and another
unique_nums.symmetric_difference_update(thisset)
```
