# range()

The range() function returns a sequence of numbers, 
starting from 0 by default, and increments by 1 (by default), and stops before a specified number.
If you do need to iterate over a sequence of numbers, the built-in function range() comes in handy. 
It generates arithmetic progressions

In many ways the object returned by range() behaves as if it is a list, but in fact it isn’t. It is an object which returns the successive items of the desired sequence when you iterate over it, but it doesn’t really make the list, thus saving space.

We say such an object is iterable, that is, suitable as a target for functions and constructs that expect something from which they can obtain successive items until the supply is exhausted. We have seen that the for statement is such a construct, while an example of a function that takes an iterable is

```python
print(range(10))
print(type(range(10)))
```

```python
nums = list(range(0, 10))
nums = list(range(0, 10, 2))

# special case, since by default behavior is increment and not decrement
nums = list(range(0, -10)) 
print(nums)

nums = list(range(0, -10, -1))
print(nums)
```

# for Loop

The for statement in Python differs a bit from what you may be used to in C or Pascal. Rather than always iterating over an arithmetic progression of numbers (like in Pascal), or giving the user the ability to define both the iteration step and halting condition (as C), Python’s for statement iterates over the items of any sequence (a list or a string), in the order that they appear in the sequence. 

```python
# Looping n times
for i in range(20):
    print(i)
    # You can also do something else
    # while you're inside the loop
```

```python
# Loopoing through the items of the list
array = ["sunyog", "suyog", "prasoon", "pratik", 22, 23, 26, 30]
for item in array:
    print(item)
```

```python
# Create a sample collection
users = {'Hans': 'active', 'Éléonore': 'inactive', '景太郎': 'active'}
for user, status in users.copy().items():
    if status == 'inactive':
        #Here we are modifying the elements of the same array that we're iterating.
        #That's why it's better to loop through the copy of array.
        #While still working on the same array. 
        del users[user]
```

To iterate over the indices of a sequence, you can combine range() and len() as follows:

```python
array = ['Mary', 'had', 'a', 'little', 'lamb']
for i in range(len(array)):
    print(array[i])
```

## for else

We can use for-else loop where the 'else' statement will be executed after the for loop has finished executing.
A loop’s else clause runs when no break occurs.

```python
nums = [1, 2, 3, 4, 5] 
  
for i in nums: 
    print(i) 
  
else: 
    print("else block is executed") 
```

```python
for i in range(2,20):
    num = i//2
    if i == 4:
        print(f"{i} is not a prime number")
        continue
    for j in range(2,num):
        if (i % j == 0):
            print(f"{i} is not a prime number")
            break
    else:
        print(f"{i} is a prime number")
    
```
