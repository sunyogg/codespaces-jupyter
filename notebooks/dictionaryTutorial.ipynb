# Dictionaries

### Creating Dictionary

```python
pairs = {"fname":"sunyog", "lname":"lodhi", "age":22, "gender":"M", "weight": 66, "occupation":"student"}
print(pairs)
```

### Using dict() to create a Dictionary

```python
#dict(_list_) -> Takes list as input, which contains nested list/set of key-value pairs
new_pairs = dict([("key", "value"), ("age", 22), (4, 16), (3, 9)])
new_pairs = dict([["key", "value"], ["age", 22], [4, 16], [3, 9]])
print(new_pairs)
```

### Accessing and Modifying values

```python
#Accessing values with the help of keys.
val = pairs["fname"]

#Accessing only keys. Returns list of keys
list_keys = pairs.keys()

#Accessing only values. Returns list of values.
list_values = pairs.values()

#Modifying value with the help of key.
pairs["weight"] = 65
```

### Adding new Key-Value pair in the dictionary

```python
#Just make sure that key is unique.
pairs["color"] = "brown"
```

### Deleting a Key-Value pair

```python
del pairs['weight']
```

### Looping through only Keys of Dictionary

```python
for key in pairs.keys():
    print(key)
```

### Looping through only Values of Dictionary

```python
for value in pairs.values():
    print(value)
```

### Looping through Key-Value pair in dictionary

```python
check_type = type(pairs.items())

for key, value in pairs.items():
    print(f"{key} -> {value}")
```

### Nested dictionary

```python
nested_pair = {"fname":"sunyog", 
               "lname":"lodhi", 
               "age":22, "gender":"M", 
               "weight": 66, 
               "hobbies": ['gaming', 'maths', 'walking', 'singing'],
               "parents":{"mother":"Kiran Lodhi", "father":"Mahesh Lodhi"}
              }
```

### Dictionary Comprehension

```python
# syntax: {key:value for key in [list_of_keys] condition_for_values}
new_pairs = {x: x**2 for x in [1, 2, 3, 4, 5, 6, 7, 8] if x%2 ==0 }
print(new_pairs)
```

### Dicitionary methods

```python
pairs = {"fname":"sunyog", "lname":"lodhi", "age":22, "gender":"M", "weight": 66, "occupation":"student"}

#Returns a copy of the dictionary
new_dict = pairs.copy()

#Returns the list of keys
pairs.keys()

#Return the list of values
pairs.values()

#Returns a list containing a tuple for each key value pair
pairs.items()

#returns the value of specified key
pairs.get("fname")

#Removes the element with the specified key
pairs.pop("weight")

#Removes the last inserted key-value pair
pairs.popitem()

#Returns the value of the specified key. If the key does not exist: insert the key, with the specified value
pairs.setdefault("lname", "patel")

#Removes all the elements from the dictionary
pairs.clear()


#Updates the dictionary with the specified key-value pairs. 
#The specified items can be a dictionary, or an iterable object with key value pairs.
new_pairs = {"name":"sunyog", "occupation":"student"}
new_pairs.update({"height":5.11, "fav_fruit":"mango"})
print(new_pairs)

#returns a dictionary with the specified keys and the specified value.
print(pairs.fromkeys(["name", "age", "color"]))
print(pairs)


```
