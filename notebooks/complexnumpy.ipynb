### Axis in numpy
- Very complex to understand (so far)
- It seems like they are define how an operation should be performed.
- Axes are defined for arrays with more than one dimension. A 2-dimensional array has two corresponding axes: the first running vertically downwards across rows (axis 0), and the second running horizontally across columns (axis 1).
- We commonly used axes when we need to aggregate, sort and manipulate.
- Axes are direction along the arrays. 



### Aggregate function using axes in multidimensional array

```python
# summing all the elements of  2 dimensional array
arr1 = np.array([
    [1, 2, 3, 8],
    [4, 6, 0, 1]
])

print(arr1.sum())
print(arr1.sum(axis = 0))
print(arr1.sum(axis = 1))
```

```python
third = np.array([[
    [7, 4, 4, 2, 0],
    [2, 8, 0, 0, 0],
    [0, 9, 4, 6, 3],
    [3, 7, 4, 6, 9]
], 

[
    [0, 9, 3, 6, 2],
    [8, 8, 5, 6, 0],
    [5, 4, 3, 6, 8],
    [2, 9, 2, 9, 1]
],
         
[
    [3, 4, 2, 1, 3],
    [5, 4, 8, 5, 4],
    [7, 5, 1, 3, 1],
    [3, 9, 9, 0, 3]
]]) 

print("Sum alonng axis 0")
print(third.sum(axis = 0), "\n\n")
# print(third.mean(axis = 0), "\n\n")

print("Sum along axis 1")
print(third.sum(axis = 1), "\n\n")
# print(third.mean(axis = 0), "\n\n")

print("Sum along axis 2")
print(third.sum(axis = 2))
# print(third.mean(axis = 0), "\n\n")
```

```python
# adding 2 3-dimensional array using axes
arr1 = np.array([
    [
        [0, 1, 2],
        [2, 3, 4],
        [4, 5, 0]
    ], 
    [
        [6, 7, 1],
        [8, 9, 3],
        [10, 0, 4]
    ]
])

arr2 = np.array([
    [
        [1, 2, 4],
        [4, 3, 2],
        [0, 5, 1]
    ], 
    [
        [3, 8, 0],
        [8, 7, 1],
        [9, 10, 3]
    ]
])


print("sum along axis = 0") # becomes depth (default)
print(np.sum((arr1, arr2), axis = 0), "\n\n")

print("sum along axis = 1") # becomes column wise
print(np.sum((arr1, arr2), axis = 1), "\n\n")

print("sum along axis = 2") # becomes row wise
print(np.sum((arr1, arr2), axis = 2), "\n\n")
```

### Joining numpy arrays
- We join SQL table on the basis of primary key, similar to that we can join 2 numpy arrays on the basis of __axes__.
- concatenate() is used to join arrays; returns a joined array. 



```python
# joining 1 dimensional array
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print(np.concatenate((arr1, arr2), axis = 0)) # axis = 1 does not exist for 1 dimension
```

```python
# joining 2 dimensional array
arr1 = np.array([
    [1, 2, 3, 8],
    [4, 6, 0, 1]
])

arr2 = np.array([
    [4, 5, 6, 7],
    [1, 3, 2, 0]
])

print("joining along axis = 0 (column)")
print(np.concatenate((arr1, arr2), axis = 0), "\n")

print("joining along axis = 1 (row)")
print(np.concatenate((arr1, arr2), axis = 1), "\n")
```

```python
# joining 3 dimensional array
# 2 columns and 1 row
arr1 = np.array([
    [
        [0, 1, 2],
        [2, 3, 4],
        [4, 5, 0]
    ], 
    [
        [6, 7, 1],
        [8, 9, 3],
        [10, 0, 4]
    ]
])

arr2 = np.array([
    [
        [1, 2, 4],
        [4, 3, 2],
        [0, 5, 1]
    ], 
    [
        [3, 8, 0],
        [8, 7, 1],
        [9, 10, 3]
    ]
])


print("join along axis = 0") # becomes depth (default)
print(np.concatenate((arr1, arr2), axis = 0), "\n\n")

print("join along axis = 1") # becomes column wise
print(np.concatenate((arr1, arr2), axis = 1), "\n\n")

print("join along axis = 2") # becomes row wise
print(np.concatenate((arr1, arr2), axis = 2), "\n\n")
```

### Stack functions
- Stacking is same as concatenation, the only difference is that stacking is done along a new axis.
- We can concatenate two 1-D arrays along the second axis which would result in putting them one over the other, ie. stacking.
- np.stack()
- np.vstack() : stacking along columns
- np.hstack() : stacking along rows.
- __np.concatenate((a, b), axis = 1) === np.hstack((a, b))__. In other words __Horizontal stacking == column wise concatenation__
- __np.concatenate((a, b), axis = 0) === np.vstack((a, b))__. In other words __vertcial stacking == depth stacking__.

```python

a = np.array([
    [
        [0, 1, 2],
        [2, 3, 4],
        [4, 5, 0]
    ], 
    [
        [6, 7, 1],
        [8, 9, 3],
        [10, 0, 4]
    ]
])

b = np.array([
    [
        [1, 2, 4],
        [4, 3, 2],
        [0, 5, 1]
    ], 
    [
        [3, 8, 0],
        [8, 7, 1],
        [9, 10, 3]
    ]
])
print("default stack")
print(np.stack((a, b)), "\n\n")

print("stack along columns")
print(np.vstack((a, b)), "\n\n")

print("stack along rows")
print(np.hstack((a, b)), "\n\n")

print("stack along depth")
print(np.dstack((a, b)))
```

### Splitting numpy arrays
- Splitting is reverse operation of Joining.
- Joining merges multiple arrays into one and Splitting breaks one array into multiple.
- __np.array_split('arr', 'num_of_splits')__
- If the array has less elements than required, it will adjust from the end accordingly. split() would fail if the array cannot be divided into subarray with all having equal number of elements.
- Similar alternates to vstack() and dstack() are available as vsplit() and dsplit().

```python
a = np.array([1, 2, 3, 4, 5, 6])
print(np.array_split(a, 3))
print(np.array_split(a, 3))
```

```python
# Splitting a 2d array
b = np.arange(12).reshape(4, 3) # a 2d array
print(b)
# print(np.array_split(b, 4))
print(np.array_split(b, 3))
```
