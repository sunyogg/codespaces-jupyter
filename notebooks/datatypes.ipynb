### Strings

```python
s = "Hello from 2026"
s1 = "12345"
s2 = "He said \"I'm gonna go home and eat.\""

#Using newline (\n) and tab operator (\t)
s3 = "\n Hello \n\t How are you?"

print(s)
print(s1)
print(s2)
print(s3)
print(type(s1))
```

### Integers

```python
num1 = 20

# Using delimeter for better reading
num2 = 1_20_40_000 

# binary number in python prefix - 0b
num3 = 0b01011101001 

print(num1)
print(num2)
print(num3)
print(type(num1))
```

### Floats

```python
f1 = 20.0345

f2 = 1e3

# The float beyond its maximum size referred as inf, Inf, INFINITY, or infinity
f3 = 23423e90090

f4 = 12_234.1_234

# converting strings and integers to float
f5 = float("234.5")
f6 = float(23)

# denoting infinity in float
f7 = float('infinity')
f8 = float('inf')
f9 = float('-inf')


print(f1)
print(type(f1))
print(f2)
print(f3)
print(f4)
print(f5)
print(f6)
print(f7)
print(type(f8))
print(type(f9))
```

### Complex number

```python
c1 = 6+4j
c2 = 2 + 3j
c3 = 1.23 + 3.67j
c4 = c1 + c2
c5 = c1*c2
print(c1)
print(c2)
print(c3)
print(c4)
print(c5)
print(type(c1))
```
