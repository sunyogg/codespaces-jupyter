# Stack plot
- Useful when pointing the compositions with time. 

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
day =     [1, 2, 3, 4, 5, 6, 7, 8, 9]

player1 = [8, 6, 5, 5, 4, 2, 1, 1, 0]
player2 = [0, 1, 2, 2, 2, 4, 4, 4, 4]
player3 = [0, 1, 1, 1, 2, 2, 3, 3, 4]

labels = ["dev1", "dev2", "dev3"]
```

```python
plt.title("Productivity of 3 Employees working on a project")
plt.xlabel("Days")
plt.ylabel("Hours")

plt.stackplot(day, player1, player2, player3, labels = labels, colors = ["#fa5252", "#12b886", "#f59f00"],
             baseline="zero")
plt.legend(loc = (0.07, 0.05))
plt.show()
plt.close()
```
