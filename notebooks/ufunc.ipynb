# numpy ufunc

- ufuncs stands for __"Universal Functions"__.
- NumPy functions that operate on the ndarray object.
- ufuncs are used to implement vectorization in NumPy which is way faster than iterating over elements.
- ufuncs also take additional arguments, like:
    - __where__ boolean array or condition defining where the operations should take place.
    - __dtype__ defining the return type of elements.
    - __out__ output array where the return value should be copied.

### What is Vectorization?
    - Converting iterative statements into a vector based operation is called vectorization.
    - It is faster as modern CPUs are optimized for such operations.

#### Creating our own universal function (ufunc)
- Create the function like we normally do.
- Then we add that function to numpy ufunc library.
    - We use __frompyfunc()__ to achieve this
    - The frompyfunc() method takes the following arguments:
        - __function__ - the name of the function.
        - __inputs__ - the number of input arguments (arrays).
        - __outputs__ - the number of output arrays.

```python
def add_this(x, y):
    result = x + y
    return result

u_add_this = np.frompyfunc(add_this, 2, 1)

one = [1, 2, 3, 4, 5]
two = [6, 7, 8, 9, 0]


# Notice how normal python operation would be performed here.
print(add_this(one, two))

# Notice how a universal function works differently compared to normal python function. 
print(u_add_this(one, two))
```

#### Check if a function is ufunc

```python
print(type(u_add_this))
print(type(np.concatenate))
```

### Mathematical operation on numpy arrays

#### Addition

```python
one = np.array([1, 2, 3, 4, 5, 6]).reshape(2, 3)
two = np.array([3, 4, 2, 1, 7, 8]).reshape(2, 3)

result = one + two # performs on axis 0
# '+' operator is similar to:
result_2 = np.add(one, two)

print(result, "\n")
print(result_2, "\n")
print(type(np.add))
```

#### Subtraction

```python
one = np.array([1, 2, 3, 4, 5, 6]).reshape(2, 3)
two = np.array([3, 4, 2, 1, 7, 8]).reshape(2, 3)

result = two - one
result_2 = np.subtract(two, one)

print(result, "\n")
print(result_2, "\n")
print(type(np.subtract))
```

#### Modulus

```python
one = np.array([1, 2, 3, 4, 5, 6]).reshape(2, 3)
two = np.array([3, 4, 2, 1, 7, 8]).reshape(2, 3)

#np.remainder() also does the same operation.
result = two % one
result_2 = np.mod(two, one)

print(result, "\n")
print(result_2, "\n")
print(type(np.mod))
```

#### summations, products and differences

```python
arr = np.arange(1, 10)
brr = np.arange(1, 10).reshape(3,3) # 2-dimensional array

print(arr, "\n")
print(brr, "\n")
```

```python
# SUMMATION returns the sum of all the elements of the array

print(arr.sum())
print(brr.sum(axis = 0))
```

```python
# CUMMULATIVE-SUM
# Cummulative sum means partially adding the elements in array.

arr.cumsum()
# 1, 2+1, 3+2+1, 4+3+2+1...
```

```python
#PRODUCTS
# returns the product of all the elements of the array. 

print(arr.prod())
print(brr.prod(axis = 1))
```

```python
# CUMMULATIVE PRODUCT
# Cummulative product means taking the product partially.

print(arr.cumprod())
# 1, 2*1, 3*2*1, 4*3*2*1...
```

```python
# DIFFERENCE
drr = np.array([1, 4, 3, 8, 2, 0])

print(np.diff(drr))
# xi = x(i+1) - xi

# You can also perform np.diff() on repead by defining 'n' attribute. 
print(np.diff(drr, n = 3))
```

#### LCM and GCD

```python
# LCM - The Lowest Common Multiple is the smallest number that is a common multiple of two numbers.

# np.lcm() will return the lcm of only 2 numbers
print(np.lcm(2, 3))
print(np.lcm(13, 17))

# To find the lcm of more than 2 numbers use reduce() function
# The reduce() method will use the ufunc, in this case the lcm() function, on each element, and reduce the array by one dimension.

arr = np.array([1, 3, 5, 7, 13, 17])

result = np.lcm.reduce(arr)

print(result)
```

```python
# GCD - Greatest Common Divisor | Greatest Common Denominator | Highest Common Factor (HCF)
# The biggest number that is a common factor of both of the numbers.

# np.gcd() will return the gcd of only 2 numbers
print(np.gcd(24, 9))

# To find the gcd of more than 2 numbers use reduce() function

print(np.gcd.reduce([120, 60, 4 ]))
```

#### Set operation in numpy
- the set arrays should only be 1-D arrays.

```python
arr = np.array([1, 1, 1, 2, 3, 4, 5, 5, 6, 7])

# We can create a 'set array' of above array by using np.unique()
u_arr = np.unique(arr)

print(u_arr)
```

```python
a = np.arange(1, 10)
b = np.arange(7, 15)

print(a)
print(b)
```

```python
# Finding the union of 2 np arrays
c = np.union1d(a, b)

print(c)
```

```python
# Finding the intersection of 2 np arrays

c = np.intersect1d(a, b)

print(c)
```

```python
# Finding the difference of 2 np arrays

c = np.setdiff1d(a, b)
# A - (A n B)
print(c)

# Note: the setdiff1d() method takes an optional argument assume_unique, which if set to True can speed up computation. 
# It should always be set to True when dealing with sets.
c = np.setdiff1d(b, a, assume_unique = True)
# B - (A n B)
print(c)
```

```python
# Finding the symmetric difference of 2 np arrays
# (A u B) - A n B)

c = np.setxor1d(a, b)
print(c)
```
