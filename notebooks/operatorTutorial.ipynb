### Plus ( + )

```python
num1 = 2
num2 = 3
num3 = num1 + num2
print(num3)

s1 = "hello"
s2 = "How are you?"
s3 = s1 + s2 # String CONCATENATION
print(s3)
```

### Subraction ( - )

```python
num1 = 1.34
num2 = 2.32
num3 = num1 - num2
print(num3)
```

```python
num = 4.32 / 2
print(num)
```

### Modulus ( % )

```python
# Returns REMAINDER
print(4 % 2)
print(10 % 4)
print(7 % 4)
```

### Exponent ( ** )

```python
# 2 to the power of 3 (2^3)
num1 = 2 ** 3
print(num1)
```

## Floor Division ( // ) 

```python
# Result is the quotient in which the digits after the decimal point are removed.
# Equivalent of int(7/2). 
#Default behavior in other low level programming languages

num1 = 7 // 2
print(num1)

num4 = -23 // 3
print(num4)
```
